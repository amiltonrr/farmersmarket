```python
gross_income_city = dataset.groupby(['City'])[['Total', 'gross income']].sum()
gross_income_city.loc['Total sales'] = gross_income_city.sum()
gross_income_city['Total profit margin'] = gross_income_city['gross income'] / gross_income_city.Total *100         
gross_income_city.sort_values(by = 'Total',ascending = False)
```
